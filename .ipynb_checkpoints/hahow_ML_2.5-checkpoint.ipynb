# matplotlib
> ### 顯示出圖片需要加上.show()

```python
from sklearn import datasets
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1,4,6,7])
plt.xlabel('xlabel')
plt.ylabel('ylabel')
```

```python
plt.plot([5,3,14,9],[33,56,73,15],'go')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
```

## axis 指定軸的範圍

```python
plt.plot([5,3,14,9],[33,56,73,45],'go')
plt.axis([1,15,30,90])
```

### 直方圖
`plt.bar(list_ex)`
### 散佈圖
`plt.scatter(list_ex)`

```python
iris = datasets.load_iris()
x = pd.DataFrame(iris['data'], columns=iris['feature_names'])
print("target name" + str(iris['target_names']))
y = pd.DataFrame(iris['target'], columns=['target'])
df = pd.concat([x,y], axis=1)
df.head()
```

```python
df.plot.scatter(x= "sepal length (cm)", y= "sepal width (cm)", c="target")
```

```python
color = {
    0:'r',
    1:'g',
    2:'b',
}
```

```python
df['color'] = df['target'].map(color)
df.head()
```

```python
df.plot.scatter(x= "sepal length (cm)", y= "sepal width (cm)", c=df.color)
```

## add plot label
ax 的應用

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.6)
plt.show()
```

# Seaborn

```python
import seaborn as sns
```

```python
sns.lmplot('sepal length (cm)','petal width (cm)', data=df, fit_reg=False, hue='target')
```

```python
target_name = {
    0:'setosa',
    1:'versicolor',
    2:'virginica'
}
df['target_name'] = df['target'].map(target_name)
```

```python
sns.lmplot('sepal length (cm)','petal width (cm)', data=df, fit_reg=False, hue='target_name')
```
